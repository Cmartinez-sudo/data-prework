```python
points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

shots_taken = len(points)
print(shots_taken)

same_points = set(points)
print(same_points)

number_of_same_points = len(same_points)
print(number_of_same_points)

#Number of same shots#
same_shot = shots_taken - number_of_same_points
print(same_shot)

#Coordinates of the points where an arrow hits another arrow#
set(x for x in points if points.count(x) > 1)

#How many arrows on each quadrant#
first_quadrant = 0
second_quadrant= 0
third_quadrant = 0
fourth_quadrant = 0
on_the_border = 0
for x, y in points:
    if (x > 0) and (y > 0):
        print("First Quadrant")
        first_quadrant = first_quadrant + 1
    elif (x > 0) and (y < 0):
        print("Fourth Quadrant")
        fourth_quadrant = fourth_quadrant + 1
    elif (x < 0) and (y > 0):
        print("Second Quadrant")
        second_quadrant = second_quadrant + 1
    elif (x < 0) and (y < 0):
        print("Third Quadrant")
        third_quadrant = third_quadrant + 1
    else:
        print("On the Border")
        on_the_border = on_the_border + 1
        
print("Nmber of arrows in first quadrant =", first_quadrant)
print("Number of arrows in second quadrant =", second_quadrant)
print("Number of arrows in third quadrant =", third_quadrant)
print("Number of arrows in fourth qaudrant =", fourth_quadrant)
print("Number of arrows on the border =", on_the_border)

#Calculate the point closest to the center#
import math

for x, y in points:
    if (x > 0 and y > 0):
        print(math.sqrt(x**2 + y**2))
    elif (x > 0 and y < 0):
        print(math.sqrt(x**2 + y**2))
    elif (x < 0 and y > 0):
        print(math.sqrt(x**2 + y**2))
    elif (x < 0 and y < 0):
        print(math.sqrt(x**2 + y**2))
    elif (x == 0 and y > 0):
        print(y)
    elif (x == 0 and y < 0):
        print(y)
    elif (x == 0 and y == 0):
        print("0")
    else:
        print(x)
        
#THe points nearest to the center are#
print("The points closest to the center are:")
print(points[1])
print(points[11])

#Defining a function for calculating distance to the center#
def distance_to_center(x, y):
    """Calculate the distance to the center"""
    print(math.sqrt((x - 0)**2 + (y - 0)**2))

#Distance to the center#
distance_to_center(0, 2)
distance_to_center(0, -2)

#The number of arrows that wont hit the target are#
for x, y in points:
    if (x >= 9 or y >= 9):
        print(True)
    else:
        print(False)
        
#There is only one arrow that does not hit the target#
print("The arrow that misses the target is: ")
print(points[20])
```
